**STUDENT MARKS ANALYZER**

A program that analyzes student marks using Python and NumPy.

```python
import numpy as np
marks=np.array([10,20,30,40,50,80,90,60,100,0,-7])
for i in marks:
  if 100>=i>=0:
    if i>85:
      print("Grade:A")
    elif 70<= i <=85:
      print("Grade:B")
    elif 50<= i <=69:
      print("Grade:C")
    else:
      print("Fail")
  else:
    print("Invalid marks")
print(f"Total marks of all students: {np.sum(marks)}")
print(f"Average marks of all students: {np.mean(marks)}")
print(f"Highest marks among all students: {np.max(marks)}")
print(f"Lowest marks among all students: {np.min(marks)}")
```

**DAILY EXPENSE TRACKER**

Program to track daily expenses and analyze spending.

```python
import numpy as np
daily_expenses=np.array([70,80,100,110])
spending_summary={
    f"Rs {daily_expenses[0]}":"Cost of Milk",
    f"Rs {daily_expenses[1]}":"Cost of Bread and butter",
    f"Rs {daily_expenses[2]}":"Cost of Fruits",
    f"Rs {daily_expenses[3]}":"Cost of Vegetables"
}
sum=np.sum(daily_expenses)
if sum>300:
  print("Exceeded daily expense limit")
print( f"Spending summary in rupees is {spending_summary}")
print(f"Total expense in rupees is:Rs {np.sum(daily_expenses)}")
print(f"Average expense in rupees is:Rs {np.mean(daily_expenses)}")
print(f"Highest expense in rupees is:Rs {np.max(daily_expenses)}")
```
